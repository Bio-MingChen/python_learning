```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
plot_global_setting = {
    'font.family': 'Comic Sans MS',
}

with plt.rc_context(plot_global_setting):
    x = [i for i in range(6)]
    y = [2*i for i in range(6)]
    plt.plot(x,y,color='red',linewidth=2,linestyle='-.',marker='^',
            markersize=10,markeredgecolor='green',markerfacecolor='blue',
            label='2x')

    plt.title('Our first graph!', fontweight='bold', y=1.05,
            color='blue',loc='center',fontdict={'fontname':'Comic Sans MS', 'fontsize':20})

    plt.xlabel('X Axis', fontsize=15, fontweight='bold', color='green')
    plt.ylabel('Y Axis', fontsize=15, fontweight='bold', color='green')

    plt.xticks([0,1,2,3,4,5],['A','B','C','D','E','F'])
    plt.yticks([0,2,4,6,8,10],['0','2','4','6','8','10'])

    plt.legend(['2x',], loc='upper right',bbox_to_anchor=(1.2,1))

    plt.tight_layout()  # 调整布局，防止内容被裁剪
    plt.show()
```
